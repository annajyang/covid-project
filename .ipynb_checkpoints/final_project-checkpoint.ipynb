```python
import pandas as pd
import numpy as np

cases_url = 'United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv' 
cases_df = pd.read_csv(cases_url)
```

```python
cases_df
```

```python
def date_checker(s):
    return s[:2] == "03" and s[-2:] == '21'

date_filter = cases_df["submission_date"].apply(date_checker)
march = cases_df[date_filter]
```

```python
by_state_by_date = march[march['state'] == 'CA'].sort_values('submission_date') # cannot use index column in a mask
by_state_by_date
```

```python
# extract only cases
month_cases = by_state_by_date['tot_cases']
month_cases
```

```python
import scipy.stats
scipy.stats.linregress([i for i in range(0, len(month_cases))], month_cases).slope
```

```python
# abbrevs = set(march['state'])
abbrevs = [ 'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY' ]
rates = {}
for state in abbrevs:
    by_state_by_date = march[march['state'] == state].sort_values('submission_date') # cannot use index column in a mask
    month_cases = by_state_by_date['tot_cases']
    if (len(month_cases) == 0): continue 
    rate = scipy.stats.linregress([i for i in range(0, len(month_cases))], month_cases).slope
    rates[abbrev_to_state[state]] = rate
#     print("slope for " + state +" : " +  str(rate))
    
print(rates)
```
