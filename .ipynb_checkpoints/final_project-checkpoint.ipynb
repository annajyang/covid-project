## QUESTION:
Is there a correlation between the case rate and vaccination rate across the 50 states in March 2021?

```python
import pandas as pd
import numpy as np

cases_url = 'United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv' 
cases_df = pd.read_csv(cases_url)
```

```python
cases_df
```

```python
def date_checker(s):
    return s[:2] == "03" and s[-2:] == '21'

date_filter = cases_df["submission_date"].apply(date_checker)
march = cases_df[date_filter]
```

```python
by_state_by_date = march[march['state'] == 'CA'].sort_values('submission_date') # cannot use index column in a mask
by_state_by_date
```

```python
# extract only cases
month_cases = by_state_by_date['tot_cases']
month_cases
```

```python
import scipy.stats
scipy.stats.linregress([i for i in range(0, len(month_cases))], month_cases).slope
```

```python
state_to_abbrev = {
  "Alabama": "AL",
  "Alaska": "AK",
  "Arizona": "AZ",
  "Arkansas": "AR",
  "California": "CA",
  "Colorado": "CO",
  "Connecticut": "CT",
  "Delaware": "DE",
  "Florida": "FL",
  "Georgia": "GA",
  "Hawaii": "HI",
  "Idaho": "ID",
  "Illinois": "IL",
  "Indiana": "IN",
  "Iowa": "IA",
  "Kansas": "KS",
  "Kentucky": "KY",
  "Louisiana": "LA",
  "Maine": "ME",
  "Maryland": "MD",
  "Massachusetts": "MA",
  "Michigan": "MI",
  "Minnesota": "MN",
  "Mississippi": "MS",
  "Missouri": "MO",
  "Montana": "MT",
  "Nebraska": "NE",
  "Nevada": "NV",
  "New Hampshire": "NH",
  "New Jersey": "NJ",
  "New Mexico": "NM",
  "New York": "NY",
  "North Carolina": "NC",
  "North Dakota": "ND",
  "Ohio": "OH",
  "Oklahoma": "OK",
  "Oregon": "OR",
  "Pennsylvania": "PA",
  "Puerto Rico": "PR",
  "Rhode Island": "RI",
  "South Carolina": "SC",
  "South Dakota": "SD",
  "Tennessee": "TN",
  "Texas": "TX",
  "Utah": "UT",
  "Vermont": "VT",
  "Virginia": "VA",
  "Washington": "WA",
  "West Virginia": "WV",
  "Wisconsin": "WI",
  "Wyoming": "WY"
}
abbrev_to_state = {}
for x,y in state_to_abbrev.items():
    abbrev_to_state[y] = x
    
```

```python
# abbrevs = set(march['state'])
abbrevs = [ 'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY' ]
rates = {}
for state in abbrevs:
    by_state_by_date = march[march['state'] == state].sort_values('submission_date') # cannot use index column in a mask
    month_cases = by_state_by_date['tot_cases']
    if (len(month_cases) == 0): continue 
    rate = scipy.stats.linregress([i for i in range(0, len(month_cases))], month_cases).slope
    rates[abbrev_to_state[state]] = rate
#     print("slope for " + state +" : " +  str(rate))
    
print(rates)
```

```python
final_cases = pd.DataFrame.from_dict(rates.items())
final_cases
```

```python
import pandas as pd
import numpy as np
import scipy

raw_data = pd.read_csv("us-daily-covid-vaccine-doses-administered.csv")

s = set()
for i, r in raw_data.iterrows():
    s.add(r["Entity"])
    
df = pd.DataFrame(columns=["State", "Vacc_Rate", "Case_Rate"])

def in_march(s):
    if s[6] == "3":
        return True
    return False

for state in s:
    data = raw_data.loc[(raw_data["Entity"] == state) & (raw_data["Day"].apply(in_march))]["daily_vaccinations"]
    #print (data.shape)
    arr = np.array(data)
    df.loc[df.shape[0]] = [state, scipy.stats.linregress(np.arange(1, arr.size+1), y=arr).slope, ""]
    #if state == "Alabama":
        #print (arr)
```

```python
final_vacc = df
final_vacc
```

```python
final_df = final_cases.merge(final_vacc, left_on = 0, right_on = 'State').drop(labels = ['Case_Rate', 0], axis = 1).rename(mapper = {1: 'Case_Rate'}, axis = 1)
final_df = final_df.set_index("State")
# axis = 1 => apply to columns
```

```python
import matplotlib.pyplot as plt

plt.scatter(final_df['Case_Rate'], final_df['Vacc_Rate'])
plt.xlabel("Daily Case Rate")
plt.ylabel("Daily Vaccination Rate")


m, b = np.polyfit(final_df['Case_Rate'], final_df['Vacc_Rate'], 1)
ind = np.arange(max(final_df['Case_Rate']))
plt.plot(ind, m*ind+ b, color = 'red')
```

```python
final_df['slopes'] = final_df['Case_Rate'] / final_df['Vacc_Rate']
final_df.sort_values('slopes')
```

## Conclusion
YES, there is a positive correlation between case rate and vaccination rate.
